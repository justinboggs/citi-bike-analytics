```python
import pandas as pd
```

```python
data1 = pd.read_csv('data/2019/JC-201901-citibike-tripdata.csv')
data2 = pd.read_csv('data/2019/JC-201902-citibike-tripdata.csv')
data3 = pd.read_csv('data/2019/JC-201903-citibike-tripdata.csv')
data4 = pd.read_csv('data/2019/JC-201904-citibike-tripdata.csv')
data5 = pd.read_csv('data/2019/JC-201905-citibike-tripdata.csv')
data6 = pd.read_csv('data/2019/JC-201906-citibike-tripdata.csv')
data7 = pd.read_csv('data/2019/JC-201907-citibike-tripdata.csv')
data8 = pd.read_csv('data/2019/JC-201908-citibike-tripdata.csv')
data9 = pd.read_csv('data/2019/JC-201909-citibike-tripdata.csv')
data10 = pd.read_csv('data/2019/JC-201910-citibike-tripdata.csv')
data11 = pd.read_csv('data/2019/JC-201911-citibike-tripdata.csv')
data12 = pd.read_csv('data/2019/JC-201912-citibike-tripdata.csv')
```

```python
append1 = data1.append(data2)
append2 = append1.append(data3)
append3 = append2.append(data4)
append4 = append3.append(data5)
append5 = append4.append(data6)
append6 = append5.append(data7)
append7 = append6.append(data8)
append8 = append7.append(data9)
append9 = append8.append(data10)
append10 = append9.append(data11)
append11 = append10.append(data12)
```

```python
append11['startdate'], append11['starttime'] = append11['starttime'].str.split(' ', 1).str
append11['stopdate'], append11['stoptime'] = append11['stoptime'].str.split(' ', 1).str
```

```python
append11['gender'].replace(to_replace=[0], value='unknown', inplace=True)
append11['gender'].replace(to_replace=[1], value='male', inplace=True)
append11['gender'].replace(to_replace=[2], value='female', inplace=True)
```

```python
append11 = append11.rename(columns = {'tripduration': 'trip_duration', 'starttime': 'start_time', 'stoptime': 'stop_time',
                                     'start station id': 'start_station_id', 'start station name': 'start_station_name',
                                     'start station latitude': 'start_station_latitude', 'start station longitude': 'start_station_longitude',
                                     'end station id': 'end_station_id', 'end station name': 'end_station_name',
                                     'end station latitude': 'end_station_latitude', 'end station longitude': 'end_station_longitude',
                                     'bikeid': 'bike_id', 'usertype': 'user_type', 'birth year': 'birth_year',
                                     'startdate': 'start_date', 'stopdate': 'stop_date'})
```

```python
append11 = append11[['trip_duration', 'start_date', 'start_time', 'stop_date', 'stop_time', 'start_station_id', 'start_station_name',
                    'start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_name', 'end_station_latitude',
                     'end_station_longitude', 'bike_id', 'user_type', 'birth_year', 'gender']]
```
